# **1) Filter size (3x3) and (5x5)**

```
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('/content/drive/MyDrive/Dataset/Rose.jpg')

blur = cv2.blur(img,(3,3))

plt.imshow(img),plt.title('Original')
plt.show()
cv2.imwrite('orig.png', img)
plt.xticks([]), plt.yticks([])
plt.imshow(blur),plt.title('Filtered with a (3x3) filter')
plt.xticks([]), plt.yticks([])
plt.show()
```

# **2) Filter size (3x3) and (5x5), stride = 2**

```
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('/content/drive/MyDrive/Dataset/Rose.jpg')
x1 = tf.keras.preprocessing.image.img_to_array(img)
x2 = tf.reshape(x1,[1,183,275,3])
kernel_in = np.ones((3,3,3,3),np.float32)/9
x = tf.constant(x2, dtype=tf.float32)
kernel = tf.constant(kernel_in, dtype=tf.float32)
y1 = tf.nn.conv2d(x, kernel, strides=[1,2,2,1], padding='VALID')
y2 = tf.reshape(y1,[91,137,3])
out = tf.keras.preprocessing.image.array_to_img(y2)
plt.title("Filtered using a (3x3) filter with stride = 2"),plt.imshow(out)
```

```
img = cv2.imread('/content/drive/MyDrive/Dataset/Rose.jpg')
x1 = tf.keras.preprocessing.image.img_to_array(img)
x2 = tf.reshape(x1,[1,183,275,3])
kernel_in = np.ones((5,5,3,3),np.float32)/25
x = tf.constant(x2, dtype=tf.float32)
kernel = tf.constant(kernel_in, dtype=tf.float32)
y1 = tf.nn.conv2d(x, kernel, strides=[1,2,2,1], padding='VALID')
y2 = tf.reshape(y1,[90,136,3])
out = tf.keras.preprocessing.image.array_to_img(y2)
plt.title("Filtered using a (5x5) filter with stride = 2"),plt.imshow(out)
```

# **3) Filter size (3x3) and (5x5), stride = 1, padding = 0**

```
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('/content/drive/MyDrive/Dataset/Rose.jpg')
x1 = tf.keras.preprocessing.image.img_to_array(img)
x2 = tf.reshape(x1,[1,183,275,3])
kernel_in = np.ones((3,3,3,3),np.float32)/9
x = tf.constant(x2, dtype=tf.float32)
kernel = tf.constant(kernel_in, dtype=tf.float32)
y1 = tf.nn.conv2d(x, kernel, strides=[1,1,1,1], padding='SAME')
y2 = tf.reshape(y1,[183,275,3])
out = tf.keras.preprocessing.image.array_to_img(y2)
plt.title("Filtered using a (3x3) filter with stride = 1 and zero padding"),plt.imshow(out)
```
