```
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5, 5, 0.01)

sig = 1 / (1 + np.exp(-x))
plt.title("Sigmoid Function")
plt.axvline(lw=1, c='black')
plt.grid(alpha=0.5, ls='-.')
plt.plot(x,sig)
```

```
tanh = 2 / (1 + np.exp(-2 * x)) -1
plt.title("TanH Function")
plt.axvline(lw=1, c='black')
plt.grid(alpha=0.5, ls='-.')
plt.plot(x,tanh)
```

```
relu = []
for i in x:
  if i > 0:
    relu.append(i)
  else:
    relu.append(0)

plt.title("ReLU Function")
plt.axvline(lw=1, c='black')
plt.grid(alpha=0.5, ls='-.')
plt.plot(x,relu)
```

```
leaky_relu = []
for i in x:
  leaky_relu.append(max(0.1 * i, i))
  
plt.title("Leaky ReLU Function")
plt.axvline(lw=1, c='black')
plt.grid(alpha=0.5, ls='-.')
plt.plot(x,leaky_relu)
```

```
elu = []
for i in x:
  if i > 0:
    elu.append(i)
  else:
    elu.append(0.5 * (np.exp(i) - 1))

plt.title("Exponential LU Function")
plt.axvline(lw=1, c='black')
plt.grid(alpha=0.5, ls='-.')
plt.plot(x,elu)
```
